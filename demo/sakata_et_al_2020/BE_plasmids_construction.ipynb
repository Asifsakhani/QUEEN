**pCMV-Target-AID construction**  
The Target-AID plasmid (pCMV-Target-AID) was constructed by assembling two fragments encoding the N- and C-terminus halves of Target-AID, which were both amplified from pcDNA3.1_pCMV-nCas-PmCDA1-ugi pH1-gRNA(HPRT) (Addgene 79620) using primer pairs RS045/HM129 and HM128/RS046, respectively, with a backbone fragment amplified from pCMV-ABE7.10 using RS047/RS048. 

```python
QUEEN(record="input/addgene_102919.gbk", product="pCMV_ABE")
QUEEN(record="input/addgene_79620.gbk",  product="pCMV_nCas_CDA1_ugi")
```

```python
description1 = "The N-terminus of Target-AID (fragment1) was amplified from pcDNA3.1_pCMV-nCas-PmCDA1-ugi pH1-gRNA(HPRT) \
(Addgene 79620) using primer pairs RS045/HM129."
QUEEN("GAGAGCCGCCACCATGGCACCGAAGAAGAAGCG",   product="RS045", process_description=description1)
QUEEN("CTGGGGCACGATATGATCCACGTCGTAGTCGGAGA", product="HM129", process_description=description1)
pCMV_nCas_CDA1_ugi.searchsequence(RS045.seq[-18:], product="FW1", process_description=description1)
pCMV_nCas_CDA1_ugi.searchsequence(HM129.seq[-18:], product="RV1", process_description=description1)
cropdna(pCMV_nCas_CDA1_ugi, FW1[0].end, RV1[0].start, product="fragment1", process_description=description1)
modifyends(fragment1, RS045.seq, HM129.rcseq, product="fragment1", process_description=description1) 
```

```python
description2 = "The C-terminus of Target-AID (fragment2) was amplified from pcDNA3.1_pCMV-nCas-PmCDA1-ugi pH1-gRNA(HPRT)\
 (Addgene 79620) using primer pairs HM128/RS046."
QUEEN(seq="CTACGACGTGGATCATATCGTGCCCCAGTCTTTTC",  product="HM128", process_description=description2)
QUEEN(seq="TTTAAACTCATTATAGCATCTTGATCTTGTTCTCTC", product="RS046", process_description=description2)
pCMV_nCas_CDA1_ugi.searchsequence(HM128.seq[-18:],  product="FW2", process_description=description2)
pCMV_nCas_CDA1_ugi.searchsequence(RS046.seq[-18:],  product="RV2", process_description=description2)
cropdna(pCMV_nCas_CDA1_ugi, FW2[0].end, RV2[0].start, product="fragment2", process_description=description2)
modifyends(fragment2, HM128.seq, RS046.rcseq, product="fragment2", process_description=description2)
```

```python
description3 = "The backbone fragment (fragment3) was amplified from pCMV-ABE7.10 using RS047/RS048"
QUEEN("ATCAAGATGCTATAATGAGTTTAAACCCGCTGATC", product="RS047", process_description=description3)
QUEEN("CTTCGGTGCCATGGTGGCGGCTCTCCCTATAG",    product="RS048", process_description=description3)
pCMV_ABE.searchsequence(RS047.seq[-18:], product="FW3", process_description=description3)[0] 
pCMV_ABE.searchsequence(RS048.seq[-18:], product="RV3", process_description=description3)[0]
cropdna(pCMV_ABE, FW3[0].end, RV3[0].start, product="fragment3", process_description=description3)
modifyends(fragment3, RS047.seq, RS048.rcseq, product="fragment3", process_description=description3)
```

```python
description4 = "The Target-AID plasmid (pCMV-Target-AID) was constructed \
by assembling the two insert fragments and the backbone fragment."
modifyends(fragment1, "*{25}/-{25}","-{28}/*{28}", product="fragment1", process_description=description4)
modifyends(fragment2, "*{28}/-{28}","-{25}/*{25}", product="fragment2", process_description=description4)
modifyends(fragment3, "*{25}/-{25}","-{25}/*{25}", product="fragment3", process_description=description4) 
joindna(fragment1, fragment2, fragment3, topology="circular", product="pCMV_Target_AID", process_description=description4)
pCMV_Target_AID.printfeature()
pCMV_Target_AID
```

**pCMV-Target-ACE construction**  
The Target-ACE plasmid (pCMV-Target-ACE) was constructed with a fragment encoding a plasmid backbone as well as ABE7.10 amplified from pCMV-ABE7.10 using RS047/RS052 and a fragment encoding the C-terminus region of Target-AID amplified from pcDNA-pCMV-nCas9 using RS051/RS046.

```python
description5 = "The backbone fragment (fragment4) was amplified from pCMV-ABE7.10 using RS047/RS052"
QUEEN(seq="ACCTCCTCCACCGTCACCCCCAAGCTGTGACA", product="RS052")
pCMV_ABE.searchsequence(RS047.seq[-18:], product="FW4", process_description=description5)
pCMV_ABE.searchsequence(RS052.seq[-18:], product="RV4", process_description=description5)[0]
cropdna(pCMV_ABE, FW4[0].end, RV4[0].start, product="fragment4", process_description=description5)
modifyends(fragment4, RS047.seq, RS052.rcseq, product="fragment4", process_description=description5)
```

```python
description6 = "The insert fragment (fragment5) encoding the C-terminus region of Target-AID was \
amplified from pcDNA-pCMV-nCas9 using RS051/RS046."
QUEEN(seq="GCTTGGGGGTGACGGTGGAGGAGGTACCGGCGG", product="RS051")
pCMV_nCas_CDA1_ugi.searchsequence(RS051.seq[-18:], product="FW5", process_description=description6)
pCMV_nCas_CDA1_ugi.searchsequence(RS046.seq[-18:], product="RV5", process_description=description6)
cropdna(pCMV_nCas_CDA1_ugi, FW5[0].end, RV5[0].start, product="fragment5", process_description=description6)
modifyends(fragment5, RS051.seq, RS046.rcseq, product="fragment5", process_description=description6)
```

```python
description7 = "The Target-ACE plasmid (pCMV-Target-ACE) was constructed by \
assembling the insert fragment and a backbone fragment."
modifyends(fragment4, "*{25}/-{25}","-{25}/*{25}", product="fragment4", process_description=description7)
modifyends(fragment5, "*{25}/-{25}","-{25}/*{25}", product="fragment5", process_description=description7)
joindna(fragment4, fragment5, topology="circular", product="pCMV_Target_ACE", process_description=description7)
```

```python
quine(pCMV_Target_ACE, output="reconstructed_genbank/pCMV-Target-ACE_reconstruction.py") 
```

**pCMV-Target-AIDmax construction**  
To construct the Target-AIDmax plasmid (pCMV-Target-AIDmax), the pUC-optimized-PmCDA1-ugi plasmid encoding the codon-optimized C-terminal region of Target-AIDmax was first constructed by the gene synthesis service of GenScript. This C terminus fragment was then amplified with primer pair SI1304/SI1307 and assembled with a nCas9 fragment amplified from pCMV-BE4max using SI1308/SI1305 and a backbone fragment amplified from pCMV-ABEmax using SI1310/SI1309. 

```python
QUEEN(record="input/addgene_112093.gbk", product="pCMV_BE4max")
QUEEN(record="input/addgene_112095.gbk", product="pCMV_ABEmax")
QUEEN(record="input/puc-optimized-pmcda1-ugi.gbk", product="opt_CDA1_ugi")
```

```python
description8 = "The fragment encoding a nCas9 fragment (fragment6) was amplified from pCMV-BE4max using SI1308/SI1305."
QUEEN(seq="ACCAAAGAAGAAGCGGAAAGTCGACAAGAAGTACAGCATCGGCCT", product="SI1308", process_description=description8)
QUEEN(seq="GTCACCTCCCAGCTGAGACAGGTCG", product="SI1305", process_description=description8)
pCMV_BE4max.searchsequence(SI1308.seq[-18:], product="FW6", process_description=description8)
pCMV_BE4max.searchsequence(SI1305.seq[-18:], product="RV6", process_description=description8)
cropdna(pCMV_BE4max, FW6[0].end, RV6[0].start, product="fragment6", process_description=description8)
modifyends(fragment6, SI1308.seq, SI1305.rcseq, product="fragment6", process_description=description8)
```

```python
description9 = "The fragment encoding the codon-optimized C-terminal region of Target-AIDmax (fragment7) \
was amplified with primer pair SI1304/SI1307."
QUEEN(seq="CCTGTCTCAGCTGGGAGGTGACGGCGGAGGAGGAACTGGAGGAGG", product="SI1304", process_description=description9)
QUEEN(seq="TCAGCGGGTTTAAACTCATTATCACAGCATTTTGATTTTGTTCTC", product="SI1307", )
opt_CDA1_ugi.searchsequence(SI1304.seq[-18:], product="FW7", process_description=description9)
opt_CDA1_ugi.searchsequence(SI1307.seq[-18:], product="RV7", process_description=description9)
cropdna(opt_CDA1_ugi, FW7[0].end, RV7[0].start, product="fragment7", process_description=description9)
modifyends(fragment7, SI1304.seq, SI1307.rcseq, product="fragment7", process_description=description9)
```

```python
description10 = "The backbone fragment (fragment8) was amplified from pCMV-ABEmax using SI1310/SI1309."
QUEEN(seq="TGATAATGAGTTTAAACCCGCTGA", product="SI1310", process_description=description10)
QUEEN(seq="GACTTTCCGCTTCTTCTTTGGTGACTCG", product="SI1309", process_description=description10)
pCMV_ABEmax.searchsequence(SI1310.seq[-18:], product="FW8", process_description=description10)
pCMV_ABEmax.searchsequence(SI1309.seq[-18:], product="RV8", process_description=description10)
cropdna(pCMV_ABEmax, FW8[0].end, RV8[0].start, product="fragment8", process_description=description10)
modifyends(fragment8, SI1310.seq, SI1309.rcseq, product="fragment8", process_description=description10)
```

```python
description11 = "The Target-AIDmax plasmid (pCMV-Target-AIDmax) was constructed \
by assembling the two insert fragments and the backbone fragment."
modifyends(fragment6, "*{22}/-{22}", "-{22}/*{22}", product="fragment6", process_description=description11) 
modifyends(fragment7, "*{22}/-{22}", "-{24}/*{24}", product="fragment7", process_description=description11) 
modifyends(fragment8, "*{24}/-{24}", "-{22}/*{22}", product="fragment8_1", process_description=description11) 
joindna(fragment6, fragment7, fragment8_1, topology="circular", product="pCMV_Target_AIDmax", process_description=description11)
```

```python
pCMV_Target_AIDmax.printfeature()
quine(pCMV_Target_AIDmax, output="reconstructed_genbank/pCMV-Target-AIDmax_reconstruction.py") 
```

**pCMV-BE4max(C) construction**  
The BE4max(C) plasmid (pCMV-BE4max(C)) was constructed to replace the C-terminal region of Target-AIDmax with the codon-optimized rAPOBEC1 and 2×UGI domains of BE4max. To this end, an nCas9 fragment obtained from pCMV-Target-AIDmax using SI447/SI1105 was assembled with rAPOBEC1 and 2×UGI fragments obtained from BE4max using SI1352/SI1357 and SI1359/SI1350, respectively, and a backbone obtained from pCMV-BE4max using SI1351/SI448. 

```python
description12 = "An nCas9 fragment (fragment9) was obtained from pCMV-Target-AIDmax using SI447/SI1105."
QUEEN(seq="GCCACATAGCAGAACTTTAAAAGTG", product="SI447", process_description=description12)
QUEEN(seq="CTTGTCATCGTCATCCTTGTA", product="SI1105", process_description=description12)
pCMV_Target_AIDmax.searchsequence(SI447.seq[-18:],  product="FW9", process_description=description12)
pCMV_Target_AIDmax.searchsequence(SI1105.seq[-18:], product="RV9", process_description=description12)
cropdna(pCMV_Target_AIDmax, FW9[0].end, RV9[0].start, product="fragment9", process_description=description12)
modifyends(fragment9, SI447.seq, SI1105.rcseq, product="fragment9", process_description=description12)
```

```python
description13 = "The rAPOBEC1 fragment (fragment10) was obtained from BE4max using SI1352/SI1357."
QUEEN(seq="GATGACGATGACAAGTCTGGCTCCTCAGAGACTGGGCCTGTCGCC", product="SI1352", process_description=description13)
QUEEN(seq="CTTCAGGCCTGTGGCCCACAGGAT", product="SI1357", process_description=description13)
pCMV_BE4max.searchsequence(SI1352.seq[-18:], product="FW10", process_description=description13)
pCMV_BE4max.searchsequence(SI1357.seq[-18:], product="RV10", process_description=description13)
cropdna(pCMV_BE4max, FW10[0].end, RV10[0].start, product="fragment10", process_description=description13)
modifyends(fragment10, SI1352.seq, SI1357.rcseq, product="fragment10", process_description=description13)
```

```python
description14 = "The 2×UGI fragment (fragment11) was obtained from BE4max using SI1359/SI1350."
QUEEN(seq="ATCCTGTGGGCCACAGGCCTGAAGACTAATCTGAGCGACATCATT", product="SI1359", process_description=description14)
QUEEN(seq="GATCAGCGGGTTTAAACTCATTATTAGACTTTCCTCTTCTTCTTG", product="SI1350", process_description=description14)
pCMV_BE4max.searchsequence(SI1359.seq[-18:], product="FW11", process_description=description14)
pCMV_BE4max.searchsequence(SI1350.seq[-18:], product="RV11", process_description=description14)
cropdna(pCMV_BE4max, FW11[0].end, RV11[0].start, product="fragment11", process_description=description14)
modifyends(fragment11, SI1359.seq, SI1350.rcseq, product="fragment11", process_description=description14)
```

```python
description15 = "The backbone fragment (fragment12) was obtained from BE4max using SI1351/SI448."
QUEEN(seq="TAATAATGAGTTTAAACCCGCTGATC", product="SI1351", process_description=description15)
QUEEN(seq="CACTTTTAAAGTTCTGCTATGTGGC", product="SI448", process_description=description15)
pCMV_BE4max.searchsequence(SI1351.seq[-18:], product="FW12", process_description=description15)
pCMV_BE4max.searchsequence(SI448.seq[-18:],  product="RV12", process_description=description15)
cropdna(pCMV_BE4max, FW12[0].end, RV12[0].start, product="fragment12", process_description=description15)
modifyends(fragment12, SI1351.seq, SI448.rcseq, product="fragment12", process_description=description15)
```

```python
description16 = "The BE4max(C) plasmid (pCMV-BE4max(C)) was constructed\
by assembling the three insert fragments and the backbone fragment."
modifyends(fragment9,"*{25}/-{25}","-{15}/*{15}",  product="fragment9",  process_description=description16)
modifyends(fragment10,"*{15}/-{15}","-{24}/*{24}", product="fragment10", process_description=description16)
modifyends(fragment11,"*{24}/-{24}","-{26}/*{26}", product="fragment11", process_description=description16)
modifyends(fragment12,"*{26}/-{26}","-{25}/*{25}", product="fragment12", process_description=description16)
joindna(fragment9, fragment10, fragment11, fragment12, topology="circular", product="pCMV_BE4maxC", process_description=description16)
```

```python
pCMV_BE4maxC.printfeature()
quine(pCMV_BE4maxC, output="reconstructed_genbank/pCMV-BE4maxC_reconstruction.py") 
```

**pCMV-ACBEmax construction**  
The ACBEmax plasmid (pCMV-ACBEmax) was constructed by assembling an ABEmax fragment obtained from pCMV-Target-ACEmax using SI447/SI1105 with the three fragments encoding the rAPOBEC1 domain, 2×UGI domain, and the two backbone fragments that were prepared to construct pCMV-BE4max(C). 

```python
description19 = "An ABEmax fragment (fragment14) obtained from pCMV-Target-ACEmax using SI447/SI1105."
pCMV_Target_ACEmax.searchsequence(SI447.seq[-18:],  product="FW14", process_description=description19)
pCMV_Target_ACEmax.searchsequence(SI1105.seq[-18:], product="RV14", process_description=description19)
cropdna(pCMV_Target_ACEmax, FW14[0].end, RV14[0].start, product="fragment14", process_description=description19)
modifyends(fragment14, SI447.seq, SI1105.rcseq, product="fragment14", process_description=description19)
```

```python
description20 = "The ACBEmax plasmid (pCMV-ACBEmax) was constructed by assembling the ABEmax fragment \
with the three fragments encoding the rAPOBEC1 domain, 2×UGI domain, and the backbone fragment."
modifyends(fragment14, "*{25}/-{25}","-{15}/*{15}", product="fragment14", process_description=description20)
joindna(fragment14, fragment10, fragment11, fragment12, topology="circular", product="pCMV_ACBEmax", process_description=description20)
```
