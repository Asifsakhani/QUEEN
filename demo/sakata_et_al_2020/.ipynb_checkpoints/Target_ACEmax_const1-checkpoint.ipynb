**pCMV-Target-ACEmax construction**  
The Target-ACEmax plasmid (pCMV-Target-ACEmax) was constructed by assembling an ABEmax fragment obtained from pCMV-ABEmax using SI945/SI1305, a fragment encoding the C-terminus region of Target-AIDmax obtained from pUC-optimized-PmCDA1-ugi using SI1304/SI1307, and a plasmid backbone obtained from pCMV-ABEmax using SI1310/SI1309. 

```python
%matplotlib inline
import sys 
sys.path.append("../../")
from dna import *
from visualize_linear_dna import *
```

```python
pCMV_ABEmax = DNA(record="input/addgene_112095.gbk", project="pCMV_ABEmax")
SI1310      = DNA(record="input/SI1310.fasta")
SI1309      = DNA(record="input/SI1309.fasta")
dest        = joindna(SI1310[0:6], pCMV_ABEmax[5843:465])
#fig1, axes = visualize(pCMV_ABEmax, 5845, 465, annotation_loc="top",with_seq=False)
#fig2, axes = visualize(pCMV_ABEmax, 0, 1000, wrap_width=100, annotation_loc="top",with_seq=True,enlarge_w=1.0)
#fig2.savefig("test.pdf",bbox_inches="tight")
#visualize(pCMV_ABEmax, 5845, 465, width=200, annotation_loc="top",with_seq=False)
```

```python
SI945      = DNA(record="input/SI945.fasta")
SI1305     = DNA(record="input/SI1305.fasta")
ins_ABEmax = pCMV_ABEmax[362:5754]
#visualize(pCMV_ABEmax,362,5754,annotation_loc="top",with_seq=False)
```

```python
pmCDA1_ugi_opt = DNA(record="input/puc-optimized-pmcda1-ugi.gb", project="opt-pmCDA1")
SI1304         = DNA(record="input/SI1304.fasta")
SI1307         = DNA(record="input/SI1307.fasta")
ins_AIDmax     = joindna(SI1304[:21], pmCDA1_ugi_opt[501:1690], flipdna(SI1307[:21]))
#visualize(pmCDA1_ugi_opt,501,1690,annotation_loc="top",with_seq=True,enlarge_w=1.0)
```

```python
product = joindna(ins_ABEmax[103:],ins_AIDmax[22:],dest[24:], topology="circular")
```

```python
product.adddnafeature(*product.finddna(SI1310)[0].sspan, feature_key="primer_bind", qualifiers={"label":"SI1310"}, strand=1)
product.adddnafeature(*product.finddna(SI1309)[0].sspan, feature_key="primer_bind", qualifiers={"label":"SI1310"}, strand=-1)
product.adddnafeature(*product.finddna(SI945)[0].sspan,  feature_key="primer_bind", qualifiers={"label":"SI945"},  strand=1)
product.adddnafeature(*product.finddna(SI1305)[0].sspan, feature_key="primer_bind", qualifiers={"label":"SI1305"}, strand=-1)
product.adddnafeature(*product.finddna(SI1304)[0].sspan, feature_key="primer_bind", qualifiers={"label":"SI1304"}, strand=1)
product.adddnafeature(*product.finddna(SI1307)[0].sspan, feature_key="primer_bind", qualifiers={"label":"SI1307"}, strand=-1)
```

```python
product.project = "Target-ACEmax"
product.writedna("pCMV-Target-ACEmax.gbk")
```
