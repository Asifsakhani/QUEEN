```python
import sys 
sys.path.append("../../")
from dna import *
```

```python
#Load source plasmid
nCas9_AID_source = DNA(record="input/addgene_79620.gbk", project="nCas9_AID_source")
pCMV_ABE         = DNA(record="input/addgene_102919.gbk",project="pCMV_ABE")
```

```python
#Crop each fragmet
frag_a = cropdna(pCMV_ABE,5257,8636)
frag_b = cropdna(nCas9_AID_source,3894,6477)
frag_c = cropdna(nCas9_AID_source,6449,9267)
```

```python
#Add overlap seqeunces
frag_a  = joindna(DNA("ATCAAGATGCTATAA"),frag_a,DNA("GCACCGAAG")) 
frag_b  = joindna(DNA("GAGAGCCGCCACC"),frag_b)
frag_c  = joindna(frag_c, DNA("TAATGAGTTTAAA"))
```

```python
#Set end seqeunces at both ends
frag_a = modifyends(frag_a,"*{25}/-{25}","-{25}/*{25}")
frag_b = modifyends(frag_b,"*{25}/-{25}","-{28}/*{28}")
frag_c = modifyends(frag_c,"*{28}/-{28}","-{25}/*{25}")
```

```python
#Join fragments
product = joindna(frag_a,frag_b,frag_c,topology="circular",project="pCMV-Target-AID")
product.writedna("pCMV-Target-AID.gbk")
product = DNA(record="pCMV-Target-AID.gbk")
exporthistory(product)
```

```python
DNA.dna_dict['nCas9_AID_source'] = DNA(seq=None, record='input/addgene_79620.gbk', project='nCas9_AID_source', topology='linear', format=None)
DNA.dna_dict['nCas9_AID_source'] = DNA(seq=None, record='input/addgene_79620.gbk', project='nCas9_AID_source', topology='linear', format=None)
DNA.dna_dict['pCMV_ABE'] = DNA(seq=None, record='input/addgene_102919.gbk', project='pCMV_ABE', topology='linear', format=None)
DNA.dna_dict['pCMV_ABE_0'] = cropdna(DNA.dna_dict['pCMV_ABE'], start=5257, end=8636, project='pCMV_ABE')
DNA.dna_dict['nCas9_AID_source_0'] = cropdna(DNA.dna_dict['nCas9_AID_source'], start=3894, end=6477, project='nCas9_AID_source')
DNA.dna_dict['dna'] = DNA(seq='ATCAAGATGCTATAA', record=None, project='dna', topology='linear', format=None)
DNA.dna_dict['nCas9_AID_source_1'] = cropdna(DNA.dna_dict['nCas9_AID_source'], start=6449, end=9267, project='nCas9_AID_source')
DNA.dna_dict['dna_0'] = DNA(seq='GCACCGAAG', record=None, project='dna', topology='linear', format=None)
DNA.dna_dict['dna_1'] = joindna(*[DNA.dna_dict['dna'],DNA.dna_dict['pCMV_ABE_0'],DNA.dna_dict['dna_0']], topology='linear', project='dna')
DNA.dna_dict['dna_2'] = DNA(seq='GAGAGCCGCCACC', record=None, project='dna', topology='linear', format=None)
DNA.dna_dict['dna_3'] = joindna(*[DNA.dna_dict['dna_2'],DNA.dna_dict['nCas9_AID_source_0']], topology='linear', project='dna')
DNA.dna_dict['dna_4'] = DNA(seq='TAATGAGTTTAAA', record=None, project='dna', topology='linear', format=None)
DNA.dna_dict['nCas9_AID_source_2'] = joindna(*[DNA.dna_dict['nCas9_AID_source_1'],DNA.dna_dict['dna_4']], topology='linear', project='nCas9_AID_source')
DNA.dna_dict['dna_7'] = modifyends(DNA.dna_dict['dna_1'], left='*************************/-------------------------', right='-------------------------/*************************', project='dna')
DNA.dna_dict['dna_10'] = modifyends(DNA.dna_dict['dna_3'], left='*************************/-------------------------', right='----------------------------/****************************', project='dna')
DNA.dna_dict['nCas9_AID_source_3'] = modifyends(DNA.dna_dict['nCas9_AID_source_2'], left='****************************/----------------------------', right='-------------------------/*************************', project='nCas9_AID_source')
DNA.dna_dict['pCMV-Target-AID_0'] = joindna(*[DNA.dna_dict['dna_7'],DNA.dna_dict['dna_10'],DNA.dna_dict['nCas9_AID_source_3']], topology='circular', project='pCMV-Target-AID')
DNA.dna_dict['pCMV-Target-AID_0'].writedna('reconstructed_pCMV-Target-AID_0.gbk')
```
